```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans

# Download NLTK stopwords corpus
nltk.download('stopwords')

nltk.download('punkt')


# Remove non-alphanumeric characters from the description column
df['description'] = df['description'].apply(lambda x: re.sub('[^a-zA-Z0-9\s]', '', x))

# Tokenize the description column using Pandas' str.split() method
df['words'] = df['description'].apply(nltk.word_tokenize)

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize the words
df['lemmatized_words'] = df['words'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

# Remove stop words from lemmatized words
df['filtered_lemmatized_words'] = df['lemmatized_words'].apply(lambda x: [word for word in x if word.lower() not in stopwords.words('english')])

# Convert lemmatized words to features using CountVectorizer
vectorizer = CountVectorizer(analyzer=lambda x: x)
df['features'] = vectorizer.fit_transform(df['filtered_lemmatized_words'].apply(lambda x: ' '.join(x))).toarray().tolist()

# K-Means clustering
kmeans = KMeans(n_clusters=10, random_state=1)
df['cluster'] = kmeans.fit_predict(df['features'].tolist())


```

```python
# Save the clustered DataFrame to a CSV file
df.to_csv('clustered_data.csv', index=False)
```
